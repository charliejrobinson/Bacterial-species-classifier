```python
predictions = np.load('/content/predictions.npy')
images = np.float32(np.load('/content/test_arrays.npy'))
real_vals = np.load('/content/real_values.npy')

real_vals = np.argmax(real_vals, axis=1)

print(real_vals)

MS11 = np.where(real_vals == 0)[0]
MX2 = np.where(real_vals == 1)[0]
SN2 = np.where(real_vals == 2)[0]

print(MS11)

sum = np.zeros(shape=(200,200))
for i in MS11:
  for j in range(1,11):
    sum += images[i,j,:,:,0]

sum /= len(MS11) / 10

plt.imshow(sum)
plt.axis('off')
plt.title('Time averaged MS11 video')
```

```python
sum2 = np.zeros(shape=(200,200))
for i in MX2:
  for j in range(1,11):
    sum2 += images[i,j,:,:,0]

sum2 /= len(MX2) / 10

plt.imshow(sum2)
plt.axis('off')
plt.title('Time averaged MX2 video')
```

```python
sum3 = np.zeros(shape=(200,200))
for i in SN2:
  for j in range(1,11):
    sum3 += images[i,j,:,:,0]

sum3 /= len(SN2) / 10

plt.imshow(sum3)
plt.axis('off')
plt.title('Time averaged SN2 video')
```
